Markers : We can use the keyword argument marker to emphasize each point with a specified marker:

Example : Mark each point with a circle:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, marker='o')
plt.show()
```

Example 2 : Mark each point with the star:

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, marker='*')
plt.show()
```

Example 3 : Mark each point with the Diamond:

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, marker='D')
plt.show()
```

Format Strings (fmt) : We can also use the shortcut string notation parameter to specify the marker.  
This parameter is also called fmt.

written with this syntax:  
marker|line|color


Example : Mark each point with a circle:

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, 'o:r')
plt.show()
```

Example 2 : Dashed Line With yellow Colour

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, 's--y')
plt.show()
```

Example 3 : Dashed/Dotted line wth green colour.

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, 'D-.g')
plt.show()
```

Marker Size : We can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

Example : Set the size of the markers to 20:

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, marker='o', ms=20)
plt.show()
```

Marker Color :   

We can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:   

You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:   

Example : Set the EDGE color to red:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

Example : Set the FACE color to red:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

Example : Set the edge color to black and face color to red :

```python
ypoint=np.array([3, 8, 1, 10])

plt.plot(ypoint, marker='o', ms=20, mec='k', mfc='r')
plt.show()
```

Example : We have too many types of colour we use any of them.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
plt.show()
```
