By default, Plot() function draw a line from point to point.  
Example : Draw a line in a diagram from position (1, 3) to position (5, 7):

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
xpoint=np.array([1, 5])
ypoint=np.array([5, 7])

plt.plot(xpoint, ypoint)
plt.show()
```

Now, we use another plot marker which is 'ring' and there notation is 'o'.   
Example : Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):

```python
 # we already import library.

xpoint=np.array([1, 8])
ypoint=np.array([3, 10])

plt.plot(xpoint, ypoint, 'o')
plt.show()
```

Multiple points.  
Example : Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

```python
xpoint=np.array([1, 2, 6, 8])
ypoint=np.array([3, 8, 1, 10])

plt.plot(xpoint, ypoint)
plt.show()
```

Default X points  

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

Example : Plotting without x-points:

```python
ypoint=np.array([3, 8, 1, 10 ,5 ,7])

plt.plot(ypoint)
plt.show()
```

Now, we recieve X-axis points are :  
The x-points in the example above are [0, 1, 2, 3, 4, 5].
